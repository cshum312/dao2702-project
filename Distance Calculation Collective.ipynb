```python
#Distance Calculation Collective (Written by Chantelle)
import pandas as pd
from pyproj import Geod

#pyproj settings
wgs84_geod = Geod(ellps='WGS84')

#import address data
amenities = pd.read_csv("latlongamenities.csv")
purchase = pd.read_csv("latlongpurchase.csv")

amenities.head(6)
```

```python
#calculate distance with coordinates
def Distance(lat1,lon1,lat2,lon2):
  az12,az21,dist = wgs84_geod.inv(lon1,lat1,lon2,lat2)
  return dist
```

```python
#df['testlat'] = geolocator.geocode("Singapore").latitude
#df['testlon'] = geolocator.geocode("Singapore").longitude

testlat = purchase['lat'].to_list()
testlon = purchase['long'].to_list()
sector = purchase['sector'].to_list()

i = 0

while i < len(testlat):
    amenities['testlat'] = testlat[i]
    amenities['testlon'] = testlon[i]
    name = "sector" + str(sector[i])
    amenities[name] = Distance(amenities['lat'],amenities['long'],amenities['testlat'],amenities['testlon'])/1000
    amenities = amenities.drop(['testlat','testlon'], axis=1)
    i += 1
    
amenities
```

```python
#convert data to csv file
amenities.to_csv('distancesamenities.csv', index=False)
```
