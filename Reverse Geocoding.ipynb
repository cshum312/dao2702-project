```python
#Reverse Geocoding (Written by Chantelle)

import googlemaps
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

#geopy settings
geolocator = Nominatim(user_agent="DAO2702")
rgeocode = RateLimiter(geolocator.reverse, min_delay_seconds=0.01)

#import address data
df = pd.read_csv("addresses.csv")
df.head(6)
```

```python
df['Coordinates'] = df['Latitude'].map(str) + ', ' + df['Longitude'].map(str)
df['Full Address'] = df['Coordinates'].apply(rgeocode)

df.head(6)
```

```python
df.to_csv('result.csv', index=False)
```
